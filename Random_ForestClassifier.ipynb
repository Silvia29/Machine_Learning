Univariate Feature Selection :

In univariate feature selection,the SelectKBest method is used that removes all but the "k highest scoring features".Each of the feature has a test score and the features with highest scores are selected. Chi-square test is used which removes features irrelavant to classification ie features on which a class is not dependent. 
https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.chi2.html#sklearn.feature_selection.chi2

Previously,using the Random Forest Classifier on the important features selected after data visualisation gives an accuracy of about 94%. Confusion Matrix shows only few wrong predictions.

Using Univariate Feature election gives an accuracy of : 96%


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

select_feature=SelectKBest(chi2,k=5).fit(x_train,y_train)

print("feature scores :",select_feature.scores_)
print("Feature columns:",x_train.columns)
```

```python
x_train2=select_feature.transform(x_train)
x_test2=select_feature.transform(x_test)

clf_2= RandomForestClassifier()
clf_2=clf_2.fit(x_train2,y_train)
acc2=accuracy_score(y_test,clf_2.predict(x_test2))
print("Accuracy is :",accuracy)

cm=confusion_matrix(y_test,clf_2.predict(x_test2))
sns.heatmap(cm,annot=True,fmt="d")
```

Recursive feature elimination with cross validation : 11 best features can be found. 
The number of features used are plotted against their cross validation scores. 
number of features vs. cross-valdation scores

```python
#Recursive feature elimination with cross validation

from sklearn.feature_selection import RFECV

clf_4=RandomForestClassifier()
rfecv=RFECV(estimator=clf_4,step=1,cv=5,scoring='accuracy')
rfecv=rfecv.fit(x_train,y_train)
print("Number of good features : ",rfecv.n_features_)
print('Best features :', x_train.columns[rfecv.support_])
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.xlabel("Number of features selected ")
plt.ylabel("Cross validation score of selected features")
plt.plot(range(1,len(rfecv.grid_scores_)+1),rfecv.grid_scores_)
plt.show()
```
